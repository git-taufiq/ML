## A simple example to run a Support Vector Machine Classifier on synthetic data

```python
from sklearn import svm
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=200, centers=2, n_features=2,
                   random_state=0, cluster_std= 1.0)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
s = svm.LinearSVC(loss='hinge',C=1)
s.fit(X,y)
```

```python
xpred = np.linspace(0,3,100)
w = s.coef_[0]
b = s.intercept_[0]
decision_boundary = -w[0]/w[1] * xpred - b/w[1]
plt.figure()
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(xpred,decision_boundary)
```

## Support Vector Machine Regression

```python
m = 50
X = 2 * np.random.rand(m, 1)
y = (4 + 3 * X + np.random.randn(m, 1)).ravel()
from sklearn.svm import LinearSVR

svm_reg = LinearSVR(epsilon=1.5)
svm_reg.fit(X, y)

x1s = np.linspace(np.min(X), np.max(X), 100).reshape(100, 1)
y_pred = svm_reg.predict(x1s)
```

```python
plt.figure()
plt.scatter(X,y)
plt.plot(x1s,y_pred)
```
